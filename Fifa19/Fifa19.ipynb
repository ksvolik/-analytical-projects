## Предобработка данных

### Обработка пропусков

Заполним пропуски в столбце "Club":

```python
players['Club'] = players['Club'].fillna('Free agent')

players['Club'].isnull().sum()
```

Посмотрим детальнее на другие пропуски в данных:

```python
players[players['Position'].isnull()].head(10)
```

Заметим, что у небольшого числа игроков (менее 1% от общего количества) отсутсвуют данные. Исключим их из данных.

```python
players = players[np.logical_not(
    players['ID'].isin(players[players['Position'].isnull()]['ID']))].reset_index(drop = True)

players.info()
```

Теперь у нас полностью заполненная таблица.

### Приведение данных к нужным типам

```python
players.head()
```

Напишем функции, чтобы в столбцах "Value" и "Wage" избавиться от лишних символов.

Для начала добавим столбец с миллионом или тысячей в зависимости от стоимости игрока, чтобы в дальнейшем умножить на это значение число, находящееся между символами в этих столбцах.

```python
# Функция, добаляющая столбец с миллионом или тысячей в зависимости от стоимости игрока
def degree (row):
    string = row['Value']
    part = re.findall('[A-z]', string)
    if part == ['M']:
        return 1000000
    elif part == ['K']:
        return 1000
    else:
        return 0    

# Функция, выделяющая число, находящееся между символами в столбце "Value"
def value (row):
    string = row['Value']
    part = re.findall('[0-9.]+', string)
    if string == '€0':
        return '0'
    else:
        return part 
    
# Функция, превращающая список в строку
def new_value (row):
    string = row['new_value']
    return ''.join(string) 
```

Сначала обработаем столбце "Value". Установим в нём стоимость игрока в миллионах евро. 

```python
players['Value_10'] = players.apply(degree, axis = 1) 

players['new_value'] = players.apply(value, axis = 1)

players['new_value'] = players.apply(new_value, axis = 1) 

players['new_value'] = players['new_value'].astype('float')

players['Value'] = players['new_value'] * players['Value_10'] / 1000000

players = players.drop(columns = ['ID', 'Value_10', 'new_value'])

players.head()
```

По аналогии обработаем столбцец "Wage", установим в нём зарплату игрока в тысячах евро в неделю:

```python
# Функция, выделяющая число, находящееся между символами в столбце "Wage"
def wage (row):
    string = row['Wage']
    part = re.findall('[0-9.]+', string)
    if string == '€0':
        return '0'
    else:
        return part 
    
# Функция, превращающая список в строку
def new_wage (row):
    string = row['new_wage']
    return ''.join(string) 
```

```python
players['new_wage'] = players.apply(wage, axis = 1)

players['new_wage'] = players.apply(new_wage, axis = 1) 

players['new_wage'] = players['new_wage'].astype('int')

players['Wage'] = players['new_wage']

players = players.drop(columns = ['new_wage'])

players.head()
```

Обработаем столбец "Height", заменим символ `'` на `.`, а затем умножим значение на 30.48, чтобы перевести футы в сантиметры:

```python
players['Height'] = players['Height'].str.replace("'", ".")

players['Height'] = players['Height'].astype('float')

players['Height'] = round(players['Height'] * 30.48)

players['Height'] = players['Height'].astype('int')

players.head()
```

Обработаем столбец "Weight", оставим только первые три цифры, а потом разделим получившееся число на 2.205, чтобы получить вес игрока в килограммах:

```python
# Функция для извлечения первых трёх цифр
def Weight (row):
    string = row['Weight']
    part = re.findall('[0-9]+', string)
    return part 

players['Weight'] = players.apply(Weight, axis = 1)

# Функция, превращающая список в строку
def new_weight (row):
    string = row['Weight']
    return ''.join(string) 

players['Weight'] = players.apply(new_weight, axis = 1) 

players['Weight'] = players['Weight'].astype('int')

players['Weight'] = round(players['Weight'] / 2.205, 2)

players.head()
```

С числовыми значениями разобрались, теперь обратим внимание на столбцы "Nationality", "Preferred Foot" и "Position". Заметим какой объём памяти нужен для хранения таблицы:

```python
players.info(memory_usage = 'deep')
```

Установим для выбранных столбцов тип данных – `category`. 

```python
players['Nationality'] = players['Nationality'].astype('category')

players['Preferred Foot'] = players['Preferred Foot'].astype('category')

players['Position'] = players['Position'].astype('category')

players.info(memory_usage = 'deep')
```

Это действие позволили уменьшить практически в два раза объём требуемой памяти для хранния таблицы, что хорошо.

### Определение топ-10 футбольных клубов

На этот вопрос также посмотрим с разных сторон. 

Определим количество игроков для каждого клуба, общую стоимость состава, суммарную зарплату игроков в неделю и средний рейтинг игроков.

```python
Clubs = players.groupby('Club', as_index = False).agg(
    {'Name': 'count', 'Value': 'sum', 'Wage': 'sum', 'Overall': 'mean'})

Clubs = Clubs.rename(
    {'Name': 'Players_count', 'Value': 'Value_sum', 'Wage': 'Wage_sum', 'Overall': 'Overall_mean'}, axis = 1)

Clubs['Overall_mean'] = Clubs['Overall_mean'].round(1)

Clubs['Value_sum'] = Clubs['Value_sum'].round(2)

Clubs['Wage_sum'] = Clubs['Wage_sum'].round(2)

Clubs.head()
```

### Портреты среднестатистических представителей разных позиций 

Для ответы на этот вопрос, сначала разберёмся с позициями игроков и разобьём всех игроков по следующим категориям:

- Вратарь
- Защитник
- Полузащитник
- Нападающий

```python
players.head()
```

**Расшифровка позиций на поле**

Вратарь:
- 'GK' – вратарь

Защитники:
- 'LB' – левый защитник
- 'RB' – правый защитник
- 'CB' – центральный защитник
- 'LCB' – левый центральный защитник
- 'RCB' – правый центральный защитник
- 'LWB' – левый фланговый защитник
- 'RWB' – правый фланговый защитник

Полузащитники:
- 'CDM' – центральный опорный полузащитник
- 'LDM' – левый опорный полузащитник
- 'RDM' – правый опорный полузащитник
- 'CM' – центральный полузащитник
- 'LCM' – левый центральный полузащитник
- 'RCM' – правый центральный полузащитник
- 'CAM' – центральный атакующий полузащитник
- 'LAM' – левый атакующий полузащитник
- 'RAM' – правый атакующий полузащитник
- 'LM' – левый полузащитник
- 'RM' – правый полузащитник

Нападающие:
- 'LW' – левый фланговый атакующий
- 'RW' – правый фланговый атакующий
- 'LF' – левый форвард
- 'RF' – правый форвард
- 'CF' – центральный форвард
- 'ST' – нападающий
- 'RS' – правый нападающий
- 'LS' – левый нападающий

Напишем функцию и поделим всех игроков на общие категории:

```python
def Position (Position):
    if Position == 'GK':
        return 'Вратарь'
    elif Position in ['LB', 'RB', 'CB', 'LCB', 'RCB', 'LWB', 'RWB']:
        return 'Защитник'
    elif Position in ['CDM', 'LDM', 'RDM', 'CM', 'LCM', 'RCM', 'CAM', 'LAM', 'RAM', 'LM', 'RM']:
        return 'Полузащитник'
    else:
        return 'Нападающий'

players['Category'] = players['Position'].apply(Position)

players['Category'] = players['Category'].astype('category')

players.head()
```

Посмотрим на распределение: 

```python
players['Category'].value_counts()
```

Для продолжения нашего исследования обработаем данные игроков с аномальным ростом. Так как таких строк достаточно много, исключать их не будет, а скорректируем рост средним для кажой позиции. 

```python
Height_GK = players.query('Height >= 160 and Category == "Вратарь"')['Height'].mean()
Height_CB = players.query('Height >= 160 and Category == "Защитник"')['Height'].mean()
Height_CM = players.query('Height >= 160 and Category == "Полузащитник"')['Height'].mean()
Height_CF = players.query('Height >= 160 and Category == "Нападающий"')['Height'].mean()

def Height (row):
    Height = row['Height']
    Category = row['Category']
    if Height < 160 and Category == 'Вратарь':
        return Height_GK
    elif Height < 160 and Category == 'Защитник':
        return Height_CB
    elif Height < 160 and Category == 'Полузащитник':
        return Height_CM
    elif Height < 160 and Category == 'Нападающий':
        return Height_CF
    else:
        return Height

players['Height'] = players.apply(Height, axis = 1)

players['Height'] = players['Height'].round().astype('int')

players.head()
```

Теперь сгруппируем данные:

```python
# Для корректного отображения таблицы поменяем формат для двух параметров
players['Nationality'] = players['Nationality'].astype('str')
players['Preferred Foot'] = players['Preferred Foot'].astype('str')

categories = players.groupby('Category').agg(
    {'Age': 'mean', 
     'Nationality': pd.Series.mode,
     'Overall': 'mean', 
     'Value': 'mean', 
     'Wage': 'mean', 
     'Preferred Foot': pd.Series.mode, 
     'Height': 'mean', 'Weight': 'mean'}).round(2);

categories
```
